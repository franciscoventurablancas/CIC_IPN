# **Colab**
Es un entorno gratuito de **Jupyter Notebook** que no requiere configuración y que se ejecuta completamente en la nube.

Tiene dos tipos de celdas: *texto* y *código*
* **Texto**: Las celdas de texto usan lenguaje simple llamado Markdown. Puedes incluir ecuaciones como en Latex. Para más información https://colab.research.google.com/notebooks/markdown_guide.ipynb
* **Código**: Código ejecutable con su respectiva salida.
